```python
import os
from os import listdir
from PIL import Image
from numpy import asarray
from numpy import expand_dims
from matplotlib import pyplot
from tensorflow.keras.models import load_model
import tensorflow as tf
import numpy as np
import pickle
import cv2
from matplotlib import pyplot as plt

from architecture import * 
from mtcnn import MTCNN
```

```python
def list_dir_visible_only(path):
    return [f for f in listdir(path) if f[0] != '.']
```

```python
#Create detector model
detector = MTCNN()

# Create the FaceNet model
face_encoder = InceptionResNetV2()
face_encoder.load_weights('facenet_keras.h5')
```

```python
folder='../../data/to_create_embeddings/'
database = {}

for filename in list_dir_visible_only(folder):
    path = folder + filename
    img = cv2.imread(folder + filename)
    img_corrected = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)          

    faces = detector.detect_faces(img_corrected)

    if len(faces)>0:
        x1, y1, width, height = faces[0]['box']         
    else:
        x1, y1, width, height = 1, 1, 10, 10
        
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    
    face = img_corrected[y1:y2, x1:x2]                        
    face = cv2.resize(face, (160, 160))
    face = asarray(face)
    
    face = face.astype('float32')
    mean, std = face.mean(), face.std()
    face = (face - mean) / std
    
    face = expand_dims(face, axis=0)
    signature = face_encoder.predict(face)
    
    database[os.path.splitext(filename)[0]] = signature

    print("Total completed signatures: ", len(database))
```

```python
myfile = open("data.pkl", "wb")
pickle.dump(database, myfile)
myfile.close()
```
